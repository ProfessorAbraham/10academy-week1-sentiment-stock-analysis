```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import talib

# Define path to your extracted CSVs
folder_path = '/content/drive/MyDrive/yfinance_data/'
csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

# Choose one file to start with
file_path = os.path.join(folder_path, csv_files[0])
df = pd.read_csv(file_path)

# Preview and clean
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
df.sort_index(inplace=True)

# Drop NA if present
df = df.dropna(subset=['Open', 'High', 'Low', 'Close', 'Volume'])

# --- Compute Technical Indicators ---

# 1. Moving Averages
df['SMA_20'] = talib.SMA(df['Close'], timeperiod=20)
df['SMA_50'] = talib.SMA(df['Close'], timeperiod=50)

# 2. RSI
df['RSI_14'] = talib.RSI(df['Close'], timeperiod=14)

# 3. MACD
df['MACD'], df['MACD_signal'], df['MACD_hist'] = talib.MACD(
    df['Close'], fastperiod=12, slowperiod=26, signalperiod=9
)

# --- Visualization ---

# 1. Closing Price with MAs
plt.figure(figsize=(14, 6))
plt.plot(df['Close'], label='Close', alpha=0.5)
plt.plot(df['SMA_20'], label='SMA 20')
plt.plot(df['SMA_50'], label='SMA 50')
plt.title('Close Price with SMA 20 & 50')
plt.legend()
plt.grid(True)
plt.show()

# 2. RSI Plot
plt.figure(figsize=(14, 4))
plt.plot(df['RSI_14'], color='purple')
plt.axhline(70, color='red', linestyle='--')
plt.axhline(30, color='green', linestyle='--')
plt.title('RSI (14)')
plt.grid(True)
plt.show()

# 3. MACD Plot
plt.figure(figsize=(14, 4))
plt.plot(df['MACD'], label='MACD', color='blue')
plt.plot(df['MACD_signal'], label='Signal', color='orange')
plt.bar(df.index, df['MACD_hist'], label='Hist', color='gray')
plt.title('MACD')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Define paths
folder_path = "/content/drive/MyDrive/yfinance_data"  # Adjust if different
output_path = "/content/drive/MyDrive/processed_stocks"
os.makedirs(output_path, exist_ok=True)

# List all CSV files
csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

# Loop through each CSV file
for file in csv_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path)

    # Basic checks
    required_cols = {'Open', 'High', 'Low', 'Close', 'Volume'}
    if not required_cols.issubset(df.columns):
        print(f"Skipping {file} due to missing columns.")
        continue

    df = df.dropna()
    if 'Date' in df.columns:
        df['Date'] = pd.to_datetime(df['Date'])
        df.set_index('Date', inplace=True)

    # Technical indicators
    df['SMA_20'] = talib.SMA(df['Close'], timeperiod=20)
    df['SMA_50'] = talib.SMA(df['Close'], timeperiod=50)
    df['RSI'] = talib.RSI(df['Close'], timeperiod=14)
    macd, macdsignal, macdhist = talib.MACD(df['Close'])
    df['MACD'] = macd
    df['MACD_Signal'] = macdsignal
    df['MACD_Hist'] = macdhist
    df['BB_upper'], df['BB_middle'], df['BB_lower'] = talib.BBANDS(df['Close'])
    df['ADX'] = talib.ADX(df['High'], df['Low'], df['Close'])
    # Visualization: Plot Close price with SMA and Bollinger Bands
    plt.figure(figsize=(14, 7))
    plt.plot(df.index, df['Close'], label='Close Price', color='blue', alpha=0.5)
    plt.plot(df.index, df['SMA_20'], label='SMA 20', color='orange')
    plt.plot(df.index, df['SMA_50'], label='SMA 50', color='purple')
    plt.plot(df.index, df['BB_upper'], label='BB Upper', linestyle='--', color='green', alpha=0.6)
    plt.plot(df.index, df['BB_middle'], label='BB Middle', linestyle='--', color='gray', alpha=0.6)
    plt.plot(df.index, df['BB_lower'], label='BB Lower', linestyle='--', color='red', alpha=0.6)
    plt.title(f'{file} - Close Price & Indicators')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Visualization: RSI
    plt.figure(figsize=(12, 3))
    plt.plot(df.index, df['RSI'], label='RSI', color='magenta')
    plt.axhline(70, color='red', linestyle='--')
    plt.axhline(30, color='green', linestyle='--')
    plt.title(f'{file} - RSI Indicator')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Visualization: MACD
    plt.figure(figsize=(12, 4))
    plt.plot(df.index, df['MACD'], label='MACD', color='blue')
    plt.plot(df.index, df['MACD_Signal'], label='Signal Line', color='orange')
    plt.bar(df.index, df['MACD_Hist'], label='MACD Histogram', color='gray')
    plt.title(f'{file} - MACD')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Save results
    df.to_csv(os.path.join(output_path, f"processed_{file}"))
    print(f"✔️ Processed and saved: {file}")
```
